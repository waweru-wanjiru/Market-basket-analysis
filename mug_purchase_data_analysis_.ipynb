```
### import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import association_rules, apriori
```

```
## import the data
data=pd.read_csv('/content/mugs_data.csv')
```

```
### view the data
data.head()
```

```
# tail of the data
data.tail()
```

```
## the number of columns and rows
data.shape
```

```
### the data info..
data.info()
```

```
### looking for null values
## the null values in memberID is non loyal customers
data.isnull().sum()
```

```
## Unique transactions
transaction_list = []
for i in data['RECEIPTID'].unique():
    tlist = list(set(data[data['RECEIPTID']==i]['DESCRIPTION']))
    if len(tlist)>0:
        transaction_list.append(tlist)
print(len(transaction_list))
```

```
# transforming the data into transactional data
x=TransactionEncoder()
y=x.fit(transaction_list).transform(transaction_list)
df=pd.DataFrame(y,columns=x.columns_)
df
```

```
### the frequency using apriori..

freq = apriori(df,min_support=0.001, use_colnames=True)
freq
```

```
#### including the number  of the items purchased per transaction
#### including the number  of the items purchased per transaction
freq['number of items']=freq['itemsets'].apply(lambda x:len(x))
freq.sort_values(by='support',ascending=0).head(20)
```

```
## converting the data into Dataframe
trans=pd.DataFrame(freq)
trans2=trans.sort_values(by='number of items',ascending=0)
```

```
## removing the forezenset
col=['itemsets']
trans2[col]=trans2[col].applymap(lambda x:tuple(x))
print(trans2.head())
```

```
## exporting the data into csv
trans2.to_csv('trans34.csv')
```

```
### calculating the rules
rules = association_rules(freq, metric='lift', min_threshold=0.1)
rules2=rules.sort_values('confidence', ascending=False)
```

```
### writing into csv
df=pd.DataFrame(rules2)
df
```

```
## removing the forezenset
cols=['antecedents','consequents',]
df[cols] = df[cols].applymap(lambda x: tuple(x))
print (df)
```

```
## exporting the data into csv
df.to_csv('rules.csv',)
```

Doing a freqency and monetary analysis to see how frequently customers buys mugs and   how much they spend

```
### drop all the null values
data=data.dropna()
data.shape
```

```
#check frequency of customer means how many transaction has been done..
frequency_df = data.copy()
frequency_df.drop_duplicates(subset=['MEMBERID','RECEIPTID'], keep="first", inplace=True) 
frequency_df = frequency_df.groupby('MEMBERID',as_index=False)['RECEIPTID'].count()
frequency_df.columns = ['MEMBERID','Frequency']
frequency_df.sort_values(by='Frequency',ascending=0)
```

```
## converting the frequency into a dataframe and exporting the data
frequency_purchase=pd.DataFrame(frequency_df)
frequency_purchase.to_csv('frequency_df.csv')
```

```
#check summed up spend of a customer with respect t

monetary_df=data.groupby('MEMBERID',as_index=False)['GROSSAMOUNT'].sum()
monetary_df.columns = ['MEMBERID','Monetary']
monetary_df.sort_values(by='Monetary',ascending=0)
```

```
### exporting the data
monetary=pd.DataFrame(monetary_df)
monetary.to_csv('monetary.csv')
```

```
#### merging both monetary and frequency
fm=frequency_df.merge(monetary_df,left_on='MEMBERID',right_on='MEMBERID')
fm.sort_values(by='Monetary',ascending=0).head(20)
```
